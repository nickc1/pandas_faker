# Bench Marking

```python
from time import time
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_faker.fake import FIELD_LIST
%matplotlib inline
```

```python
d = {}
for field in FIELD_LIST:
    print(field)
    
    F = PandasFaker([field])
    start = time()
    
    X = F.make_fakes(1000)
    
    d[field] = time() - start
    
    
    
```

```python
len(X)
```

```python
fig, ax = plt.subplots(figsize=(8,25))
pd.Series(d).sort_values().plot(kind='barh', ax=ax)

ax.set_title('Time to Generate 1000 Samples')
ax.set_xlabel('Time (Seconds)')
```

```python
fig.savefig('time_to_gen_1k_samples.png', bbox_inches='tight', dpi=150)
```
